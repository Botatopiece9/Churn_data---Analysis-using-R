```r
library(dplyr)
library(readr)
library(tidyr)

#reading the data
raw <- read_csv("raw_churn_data.txt", col_types = cols(.default = col_character()))

#replacing bad string with NA
stat_mode <- function(x) {
  x <- x[!is.na(x)]
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}

df <- raw %>%
  mutate(
    across(everything(), ~ na_if(trimws(.x), "")),  # blank → NA
    across(everything(), ~ na_if(.x, "unknown"))    # "unknown" → NA
  )

#converting columns to correct types
df <- df %>%
  mutate(
    Age            = as.numeric(Age),
    TenureMonths   = as.numeric(TenureMonths),
    MonthlyCharges = as.numeric(MonthlyCharges),
    TotalCharges   = as.numeric(TotalCharges),
    Churn          = as.integer(Churn)
  )

#fixing inconsistencies
df <- df %>%
  mutate(TotalCharges = if_else(TotalCharges < 0, NA_real_, TotalCharges))
df <- df %>%
  mutate(
    InternetService = tools::toTitleCase(tolower(InternetService)),
    ContractType    = tools::toTitleCase(tolower(ContractType)),
    PaymentMethod   = tools::toTitleCase(tolower(PaymentMethod))
  )

#account for missing values
df <- df %>%
  mutate(
    Age            = if_else(is.na(Age),            median(Age,            na.rm = TRUE), Age),
    TenureMonths   = if_else(is.na(TenureMonths),   median(TenureMonths,   na.rm = TRUE), TenureMonths),
    MonthlyCharges = if_else(is.na(MonthlyCharges), median(MonthlyCharges, na.rm = TRUE), MonthlyCharges),
    TotalCharges   = if_else(is.na(TotalCharges),   median(TotalCharges,   na.rm = TRUE), TotalCharges),
    ContractType    = if_else(is.na(ContractType),    stat_mode(ContractType),    ContractType),
    InternetService = if_else(is.na(InternetService), stat_mode(InternetService), InternetService),
    PaymentMethod   = if_else(is.na(PaymentMethod),   stat_mode(PaymentMethod),   PaymentMethod)
  )

#final clean
df <- df %>%
  select(-CustomerID) %>%
  mutate(
    ContractType    = factor(ContractType),
    InternetService = factor(InternetService),
    PaymentMethod   = factor(PaymentMethod),
    Churn           = factor(Churn, levels = c(0, 1), labels = c("No", "Yes"))
)
  
write.csv(df,"cleaned_churn_data.csv", row.names = FALSE)
```

```r
#training and testing
train_idx  <- createDataPartition(df$Churn, p = 0.8, list = FALSE)
train_data <- df[ train_idx, ]
test_data  <- df[-train_idx, ]

#logistic regression
ctrl <- trainControl(
  method          = "cv",
  number          = 5,
  classProbs      = TRUE,
  summaryFunction = twoClassSummary,
  savePredictions = "final"
)

model <- train(
  Churn ~ .,
  data      = train_data,
  method    = "glm",
  family    = "binomial",
  trControl = ctrl,
  metric    = "ROC"
)

model
```

```r
#test set evaluation
pred_probs <- predict(model, newdata = test_data, type = "prob")[, "Yes"]
pred_class <- predict(model, newdata = test_data)

confusionMatrix(pred_class, test_data$Churn, positive = "Yes")

#ROC curve
roc_obj <- roc(test_data$Churn, pred_probs, levels = c("No", "Yes"))

roc_df <- data.frame(
  FPR = 1 - roc_obj$specificities,
  TPR = roc_obj$sensitivities
)

ggplot(roc_df, aes(x = FPR, y = TPR)) +
  geom_line(color = "#2196F3", linewidth = 1.2) +
  geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "gray50") +
  annotate("text", x = 0.65, y = 0.15,
           label = paste0("AUC = ", round(auc(roc_obj), 3)), size = 5) +
  labs(title = "ROC Curve", x = "False Positive Rate", y = "True Positive Rate") +
  theme_minimal()

png("roc_curve.png")
plot(roc_obj, main="ROC Curve")
dev.off()

#PR curve
actual_binary <- as.integer(test_data$Churn == "Yes")
pr_obj <- pr.curve(
  scores.class0 = pred_probs[actual_binary == 1],
  scores.class1 = pred_probs[actual_binary == 0],
  curve = TRUE
)

pr_df <- data.frame(
  Recall    = pr_obj$curve[, 1],
  Precision = pr_obj$curve[, 2]
)

ggplot(pr_df, aes(x = Recall, y = Precision)) +
  geom_line(color = "#E91E63", linewidth = 1.2) +
  annotate("text", x = 0.2, y = 0.15,
           label = paste0("AUC-PR = ", round(pr_obj$auc.integral, 3)), size = 5) +
  labs(title = "Precision-Recall Curve", x = "Recall", y = "Precision") +
  ylim(0, 1) + theme_minimal()

png("pr_curve.png")
plot(pr_obj, main="PR Curve")
dev.off()
```
